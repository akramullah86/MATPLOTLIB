```python
batsman = pd.read_csv('E:/Datasets/Matplotlib_datasets/sharma-kohli.csv')
```

```python
x = batsman['index']
y = batsman['V Kohli']
z = batsman['RG Sharma']
plt.plot(x,y)
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

```python
# labels title
plt.title('BATSMAN SCORE COMPARISON',color='Blue')
plt.plot(x,y,color='darkred')
plt.plot(x,z,color='blue')
```

```python
plt.plot(x,y,color='blue')
plt.plot(x,z,color='brown')
plt.title('SHARMA AND KOHLI SCORE COMAPRISON',color='blue')
plt.xlabel('YEAR',color='blue')
plt.ylabel('RUNS',color='blue')
```

```python
plt.title('SCORE COMPARISON OF BATSMAN',size='15',color='blue')
plt.plot(x,y,color='black',linestyle='solid',linewidth=3)
plt.plot(x,z,color='red',linestyle='solid',linewidth=3)
plt.xlabel('YEAR',size=20,color='blue')
plt.ylabel('RUNS SCORED',size = 20,color='blue')
```

```python
plt.plot(x,y,'o:r')
plt.plot(x,z,'--')
```

```python
plt.plot(x,y,color='r',linestyle='dotted',linewidth=3,marker='o',markersize=10)
plt.plot(x,z,color='y',linestyle='dotted',linewidth=3,marker='o',markersize=10)

plt.title('SCORE COMPARISON OF BATSMAN',size='15',color='blue')
plt.xlabel('YEAR',size=20,color='blue')
plt.ylabel('RUNS SCORED',size = 20,color='blue')
```

```python
plt.plot(x,y,color='y',linestyle='solid',linewidth=3,marker='*',markersize=10,label='Kohli')
plt.plot(x,z,color='b',linestyle='solid',linewidth=3,marker='*',markersize=10,label='Sharma')

plt.title('SCORE COMPARISON OF BATSMAN',color='b',size=15)
plt.xlabel('SEASON',size='15',color='b')
plt.ylabel('RUNS SCORED',size=15,color='b')
plt.legend()
```

```python
plt.plot(x,y,color='y',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Kohli')
plt.plot(x,z,color='b',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Sharma')

plt.title('SCORE COMPARISON OF BATSMAN',color='b',size=15)
plt.xlabel('SEASON',size='15',color='b')
plt.ylabel('RUNS SCORED',size=15,color='b')
plt.legend(loc='upper right')
```

```python
plt.plot(x,y,c='b',linestyle='-',linewidth=3,marker='o',markersize=15,label='Kohli',mfc='r')
plt.plot(x,z,c='r',linestyle='-',linewidth=3,marker='o',markersize=15,label='Sharma',mfc='b')

plt.title('SCORE COMPARISON OF BATSMAN',color='b',size=15)
plt.xlabel('SEASON',size='15',color='b')
plt.ylabel('RUNS SCORED',size=15,color='b')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
