## BAR CHART 

`Use for numerical to categorical data analysis`

`Use for Bivariat analysis`

```python
students = [10,20,40,30,60,15]
color = ['red','green','blue','purple','pink','black']
```

```python
plt.bar(color,students,color='darkblue')
```

```python
# draw horizontal bar graph

plt.barh(color,students,color='darkred')
plt.show()
```

```python
students = [10,20,40,30,60,15]
color = ['redrederederedredredererdredredred','greengreengreeengreen','blueblueblueblueblueblue','purplepurplepurplepurple','pink','black']
```
